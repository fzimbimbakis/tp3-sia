```python
from multilayer_perceptron import MultilayerPerceptron
from fonts import get_fonts, print_letter
import numpy as np

# BATCH y eta adaptativo
```

```python
fonts = get_fonts()
layers = np.array([fonts[0].size, 20, 10, 2, 10, 20, fonts[0].size])
learning_rate = 0.0005
epochs = 20000

perceptron = MultilayerPerceptron(fonts, fonts, learning_rate, layers, batch_size=fonts[0].size, learning_rate_params=[0.0005, 0.2, 20])

perceptron.train(epochs)
```

```python
output = perceptron.test_input(fonts)
for text, i in enumerate(fonts):
    print_letter(i)
    print(fonts[text])
    print(output[text])


```
