```python
from multilayer_perceptron import MultilayerPerceptron
from fonts import get_fonts, print_letter, font_header
import numpy as np
import matplotlib.pyplot as plt

# BATCH y eta adaptativo
```

```python
fonts = get_fonts()
layers = np.array([fonts[0].size, 20, 2, 20, fonts[0].size])
learning_rate = 0.05
epochs = 25000

perceptron = MultilayerPerceptron(fonts, fonts, learning_rate, layers, batch_size=fonts[0].size, learning_rate_params=[0.005, 0.2, 20])

perceptron.train(epochs)
```

```python
output = perceptron.test_input(fonts)
for text, i in enumerate(fonts):
    print_letter(i)
    print(fonts[text])
    print_letter(output[text])
    print(output[text])
    print(" ")
    print(" ")


```

```python
for o in output:
    o = np.array(o)
    plt.imshow(o.reshape(7,5), cmap='binary', interpolation='nearest')
    plt.show()
```

```python
encoded_inputs = perceptron.encode(fonts)

labels=np.array(font_header)

x, y = zip(*encoded_inputs)
plt.scatter(x, y)
for i, text in enumerate(labels):
    plt.annotate(text, (x[i], y[i]))
plt.show()
```
